```python
# Re-importing necessary libraries due to the state reset
import numpy as np
from scipy.stats import lognorm
import matplotlib.pyplot as plt
import japanize_matplotlib

# First graph: Log-normal distribution with different parameters
x = np.linspace(0.01, 5, 1000)

plt.figure(figsize=(10, 6))
params = [(0.5, 0), (1, 0), (1.5, 0), (2, 0)]  # (shape, location)
for shape, loc in params:
    pdf = lognorm.pdf(x, s=shape, loc=loc, scale=np.exp(loc))
    plt.plot(x, pdf, label=f"sigma={shape}, mu={loc}")
plt.title("様々なパラメタのもとでの対数正規分布")
plt.xlabel("x")
plt.ylabel("確率密度関数")
plt.legend()
plt.grid()
plt.savefig('lognorm1.png')
plt.show()

# Second graph: Median and mean of a specific log-normal distribution
shape = 1  # Standard deviation of log(X)
loc = 0    # Location parameter (log scale)
scale = 1  # Scale parameter (exp(loc))

pdf = lognorm.pdf(x, s=shape, loc=loc, scale=scale)
median = lognorm.median(s=shape, loc=loc, scale=scale)
mean = lognorm.mean(s=shape, loc=loc, scale=scale)

plt.figure(figsize=(10, 6))
plt.plot(x, pdf, label="Log-normal PDF", color="blue")
plt.axvline(median, color="green", linestyle="--", label=f"Median = {median:.2f}")
plt.axvline(mean, color="red", linestyle="--", label=f"Mean = {mean:.2f}")
plt.title("Log-normal Distribution with Median and Mean")
plt.xlabel("x")
plt.ylabel("Probability Density Function (PDF)")
plt.legend()
plt.grid()
plt.show()
```
